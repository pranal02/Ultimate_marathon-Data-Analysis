```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("TWO_CENTURIES_OF_UM_RACES.csv")
```

```python
df.head(5)
```

```python
df.shape
```

```python
df.dtypes
```

```python
#clean up data 
# only wants USA Races , 50k or 50 mile , 2020
##step 1 show 50 miles or 50 km
df[df['Event distance/length']=='50km']
```

```python
df[df['Event distance/length']=='50mi']
```

```python
#But we want to work on with both 50km and 50 mile and of specific year =2020
df[df['Event distance/length'].isin(['50km','50mi']) & (df['Year of event']==2020)]
```

```python
# we are done with year and length of the race now calculate the country = USA
df[df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0)== 'USA']
```

```python
#combine all filters together to work on the analysis of Race/marathon eveents
df2 = df[(df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0)== 'USA') & (df['Event distance/length'].isin(['50km','50mi']) & (df['Year of event']==2020))]
```

```python
df2.head(5)
```

```python
df2.shape
```

```python
# remove country name from Event name
df2['Event name']=df2['Event name'].str.split('(').str.get(0)
```

```python
df2.head(3)
```

```python
# remove h from performance hour
df2['Athlete performance']=df2['Athlete performance'].str.split(' ').str.get(0)
```

```python
# drop the columns  athlete club , country , year of birth , age category

df2=df2.drop(['Athlete club','Athlete country','Athlete year of birth','Athlete age category'],axis=1)
```

```python
# clean up null values
df2.isna().sum()
```

```python
# to drop the null values we can use df2.dropna()
```

```python
#check for duplicates
df2[df2.duplicated()==True]
```

```python
#reset index
df2.reset_index(drop=True)
```

```python
#make avaerage speed as float data type
df2['Athlete average speed']=df2['Athlete average speed'].astype(float)
```

```python
# rename s columns
df2 = df2.rename(columns={'Year of event':'year', 'Event dates': 'race_day','Event name': 'race_name','Event distance/length': 'race_length','Event number of finishers': 'race_number_of_finishers','Athlete performance': 'athlete_performance','Athlete gender': 'athlete_gender','Athlete average speed': 'athlete_average_speed','Athlete ID': 'athlete_id',})
```

```python
# SEasons for the data -> slower in summer than Winter?
#spring 3-5
#summer 6-8
#fall 9-11
#winder 12-2
#splite between two decimals
```

```python
df2['race_month'] = df2['race_day'].str.split('.').str.get(1).astype(int)
df2.head()
```

```python
df2['race_season']=df2['race_month'].apply(lambda x : 'Winter' if x >11 else 'Fall' if x>8 else 'Summer' if x >5  else 'Spring'  if x> 2 else 'Winter' )
```

```python
df2.head()
```

```python
df2.groupby('race_season')['athlete_average_speed'].agg(['mean','count']).sort_values('mean',ascending=False)
```
